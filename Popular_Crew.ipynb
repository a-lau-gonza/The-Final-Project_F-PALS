```python
import pandas as pd
import requests
from API_Key import api_key
titleNid = pd.read_csv("titleNid.csv")
titleNid
```

```python
famous_actors = pd.DataFrame()
id_for_movie = []

for movie_id in titleNid['id']:
    movie = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={api_key}&language=en-US').json()
    for popular in movie['cast']:
        if popular["popularity"] >=10:
            famous_actors = famous_actors.append(popular,ignore_index = True)
            id_for_movie.append(movie_id)
            print(id_for_movie)
```

```python
famous_actors["movie_id"] = id_for_movie
famous_actors
```

```python
directors = pd.DataFrame()
director_movie_id = []

for movie_id in titleNid['id']:
    movie  = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={api_key}&language=en-US').json()
    for crew_member in movie['crew']:
        if crew_member['job'] == "Director":
            directors = directors.append(crew_member, ignore_index=True)
            director_movie_id.append(movie_id)
            
```

```python
directors['movie_id'] = director_movie_id
directors
```
