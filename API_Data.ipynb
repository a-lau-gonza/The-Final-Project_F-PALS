```python
import requests
import pandas as pd
from API_Key import api_key
```

```python
pages = list(range(1,11))

popular = pd.DataFrame()
```

```python
for page in pages:
    movieData = pd.DataFrame(requests.get(f'https://api.themoviedb.org/3/movie/popular?api_key={api_key}&language=en-US&page={page}').json()['results'])
    popular = popular.append(movieData)

popular.head()
```

```python
popular = popular[popular['adult'] == False]
popular_features = popular.drop(['adult', 'backdrop_path', 'id', 'original_title', 'overview', 'poster_path', 'title','video'],1)
popular_features.head()
```

```python
titleNid = popular[['id','title']]
titleNid
```

```python
genre = requests.get(f'https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}&language=en-US').json()['genres']
genre_df = pd.DataFrame(genre)
genre_df
```

```python
test = requests.get(f'https://api.themoviedb.org/3/movie/370172/credits?api_key={api_key}&language=en-US').json()
dftest = pd.DataFrame()
for popular in test['cast']:
    if popular["popularity"] >=25:
        dftest = dftest.append(popular,ignore_index = True)
dftest.head()
```
